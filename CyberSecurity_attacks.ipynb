# *Data Processing*

```python
import warnings
warnings.filterwarnings("ignore")

df_filled = df.fillna(df.mean())
df_filled = pd.get_dummies(df_filled, columns=['Geo-location Data'], drop_first=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_filled[['Packet Length', 'Anomaly Scores']])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Use the elbow method to find the optimal number of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```


The elbow method is a technique used to find the optimal number of clusters (k) in a dataset. The idea is to plot the explained variation (or inertia) as a function of the number of clusters and look for an "elbow" point in the plot where the rate of decrease in inertia slows down. The point where this slowdown occurs is often considered the optimal number of clusters.IN this case, it would be 4.

```python
import seaborn as sns

# Apply k-means clustering with optimal k value of 4
kmeans = KMeans(n_clusters=4, random_state=42)
df_filled['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Packet Length', y='Anomaly Scores', hue='Cluster', data=df_filled, palette='viridis')
plt.title('Clustering of Attacks')
plt.xlabel('Packet Length (Standardized)')
plt.ylabel('Anomaly Scores (Standardized)')
plt.show()
```
